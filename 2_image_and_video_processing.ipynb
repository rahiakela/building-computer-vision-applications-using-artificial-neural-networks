## Loading, Exploring, and Showing an Image

```
from __future__ import print_function
import cv2
from google.colab.patches import cv2_imshow
```

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# image is a NumPy array
print("Dimensions of the image: ", image.ndim)
print("Image height: ", format(image.shape[0]))
print("Image width: ", format(image.shape[1]))
print("Image channels: ", format(image.shape[2]))
print("Size of the image array: ", image.size)

# Display the image
cv2_imshow(image)
```

The image NumPy array consists of three dimensions: height  width  channel. The
first element of the array is the height, which tells us how many rows our pixel grid has.

## OpenCV Code to Access and Manipulate Pixels

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# Access pixel at (0,0) location 
(b, g, r) = image[0, 0]
print("Blue, Green and Red values at (0, 0): ", (b, g, r))

# Manipulate pixels and show modified image
image[0:100, 0:100] = (255, 255, 0)

# Display the image
cv2_imshow(image)
```

We are taking a range of pixels from 0 to 100 along the y-axis and from 0
to 100 along the x-axis to form a 100100 square and assigning the values (255,255,0) or pure blue, pure green, and no red to all the pixels within this square.

## Drawing

OpenCV provides convenient methods to draw shapes on an image. We will learn how to draw a line, rectangle, and circle on an image using the following methods:

- Line: cv2.line()
- Rectangle: cv2.rectangle()
- Circle: cv2.circle()

### Drawing a Line on an Image

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# 1-Decide the coordinates of the starting/ending position of the line
start = (0, 0)  # set start and end coordinates(location (0,0) is the top-left corner of the image)
end = (image.shape[1], image.shape[0])  # it represents the coordinates of the bottom-right corner of the image.

# 2-Set the color of the line.
color = (255, 0, 0)  # set the color in BGR(blue)

# 3-Optionally, set the thickness of the line.
thickness = 4  # set thickness in pixel

# You have probably guessed by now that we are drawing a diagonal line.
cv2.line(image, start, end, color, thickness)

# Display the image
cv2_imshow(image)
```

### Drawing a Rectangle on an Image

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# 1-Decide the coordinates of the starting/ending position of the line
start = (100, 70)  # set start and end coordinates of the top-left and bottom-right corners of the rectangle
end = (350, 380)  # it represents the coordinates of the bottom-right corner of the image.

# 2-Set the color and thickness of the outline
color = (0, 255, 0)  # set the color in BGR(blue)

# 3-Optionally, set the thickness of the line.
thickness = 5  # set thickness in pixel

# Draw the rectangle
cv2.rectangle(image, start, end, color, thickness)

# Save the modified image with the rectangle drawn to it.
cv2.imwrite("rectangle.jpg", image)

# Display the image
cv2_imshow(image)
```

We will now slightly modify this example and draw the rectangle on a blank canvas. We will first create a canvas (as opposed to loading an existing image) and draw a rectangle on it. We will then save and display the resultant image.

```
import numpy as np
```

```
# create a new canvas
canvas = np.zeros((200, 200, 3), dtype="uint8")

start = (10, 10)  
end = (100, 100)  

color = (0, 0, 255)  
thickness = 5

# Draw the rectangle
cv2.rectangle(canvas, start, end, color, thickness)

# Save the modified image with the rectangle drawn to it.
cv2.imwrite("canvas-rectangle.jpg", canvas)

# Display the image
cv2_imshow(canvas)
```

```
# create a new canvas
canvas = np.zeros((200, 200, 3), dtype="uint8")

start = (10, 10)  
end = (100, 100)  

color = (0, 0, 255)  
thickness = -1

# Draw the rectangle
cv2.rectangle(canvas, start, end, color, thickness)

# Save the modified image with the rectangle drawn to it.
cv2.imwrite("canvas-rectangle.jpg", canvas)

# Display the image
cv2_imshow(canvas)
```

### Drawing a Circle on an Image

Drawing a circle on an image is equally easy. You create your own canvas or load an existing image and then set the coordinates of the center, radius, color, and thickness of the outline of the circle.

```
# create a new canvas
canvas = np.zeros((200, 200, 3), dtype="uint8")
 
center = (100, 100) # The coordinates of the center of the circle.
radius = 50         # The radius of the circle.
color = (0, 0, 255) # The color of the outline of the circle.
thickness = 5       # The thickness of the outline.

# Draw the rectangle
cv2.circle(canvas, center, radius, color, thickness)

# Save the modified image with the rectangle drawn to it.
cv2.imwrite("circle.jpg", canvas)

# Display the image
cv2_imshow(canvas)
```

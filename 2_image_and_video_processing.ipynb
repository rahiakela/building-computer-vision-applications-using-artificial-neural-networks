## Loading, Exploring, and Showing an Image

```
from __future__ import print_function
import cv2
from google.colab.patches import cv2_imshow
```

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# image is a NumPy array
print("Dimensions of the image: ", image.ndim)
print("Image height: ", format(image.shape[0]))
print("Image width: ", format(image.shape[1]))
print("Image channels: ", format(image.shape[2]))
print("Size of the image array: ", image.size)

# Display the image
cv2_imshow(image)
```

```
print("Size of the image array: ", image.size)
print("Size of the image array: ", image.shape[0] * image.shape[1] * image.shape[2])
```

The image NumPy array consists of three dimensions: height × width × channel. The
first element of the array is the height, which tells us how many rows our pixel grid has.

## OpenCV Code to Access and Manipulate Pixels

```
# image path
image_path = "marsrover.png"
# Read or load image from its path
image = cv2.imread(image_path)

# Access pixel at (0,0) location 
(b, g, r) = image[0, 0]
print("Blue, Green and Red values at (0, 0): ", (b, g, r))

# Manipulate pixels and show modified image
image[0:100, 0:100] = (255, 255, 0)

# Display the image
cv2_imshow(image)
```

We are taking a range of pixels from 0 to 100 along the y-axis and from 0
to 100 along the x-axis to form a 100×100 square and assigning the values (255,255,0) or pure blue, pure green, and no red to all the pixels within this square.
